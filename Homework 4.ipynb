```python
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1,2,3,4,5,6])
ypoints = np.array([5,6,7,8,9,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xplots = np.array([3,6,9,12])
yplots = np.array([2,8,1,10])

plt.plot(xplots, yplots, marker = 's')
plt.show()
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([2,4,6,14,10,12])

plt.plot(x, y, marker = 's', ms = 10, mec = 'g', mfc = 'g',
        linestyle = '--', color = 'r')
plt.show()
```

```python
x1 = np.array([0,4])
y1 = np.array([1,5])
y2 = np.array([5,9])
y3 = np.array([9,13])

plt.plot(x1,y1)
plt.plot(x1,y2)
plt.plot(x1,y3)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
```

```python
names = list(marks.keys())
grades = list(marks.values())

plt.bar(range(len(marks)), grades, tick_label = names)
plt.title('Student Marks')
plt.show()

plt.pie(grades, labels = names)
plt.legend(bbox_to_anchor = (1,1), loc = 'upper left')
plt.show()
```

```python
#plot lines (1,1)
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = np.array([1,2,3,4])
d = np.array([4,3,2,1])
e = np.array([2,3,4,5])
f = np.array([1,2,5,6])
plt.subplot(2,3,1)
plt.plot(a,b,c,d,e,f)
plt.title('LINES')

#plot bar(1,2)
fruits = ['a','b','c','d','e']
amount = [3,6,4,7,9]
plt.subplot(2,3,2)
plt.bar(fruits, amount)
plt.title('BAR')

#plot pie (1,3)
car = ['sedan', 'suv','van,','truck']
ontheroad =[2,4,6,8]
plt.subplot(2,3,3)
plt.pie(ontheroad, labels = car)
plt.title('PIE')

#plot grid (2,1)
g = np.array([2,4,3,8])
h = np.array([1,2,4,4])
plt.subplot(2,3,4)
plt.grid(color = 'green')
plt.plot(g,h)
plt.title('GRID')

#plot scatter (2,2)
i = np.array([1,3,9,2,1,3,4,5])
j = np.array([8,3,7,6,5,0,9,6])
plt.subplot(2,3,5)
plt.scatter(i,j)
plt.title('SCATTER')

#plot histogram (2,3)
l = np.random.normal(40,20,70)
plt.subplot(2,3,6)
plt.hist(l)
plt.title('HISTOGRAM')

plt.suptitle('SUPER TITLE')
plt.show()
```
